```python
import pandas as pd
from snakemake.io import expand
from sm_utils import *
import os
import yaml
```

```python
def load_config(config_file=None):
    """
    Load snakemake config file as a dictionary
    """
    if not config_file:
        # d = os.path.dirname(__file__)
        od = f'../snakemake/'
        config_file = f'{od}/config.yml'

    with open(config_file) as f:
        config = yaml.safe_load(f)

    return config
```

```python
df = parse_config('config.tsv')
config = load_config()
```

```python
def get_df_val(df, col, col2_val, col2='sample'):
    temp = df.loc[df[col2]==col2_val]
    assert len(temp.index) == 1
    return temp[col].values[0]
```

```python
tech_rep = 'GM22300_1'
expand(config['iq']['gtf'],lab_rep=get_df_val(df, 'lab_rep', tech_rep, 'tech_rep'))
```

```python
fname = 'config.tsv'
```

```python
df = pd.read_csv(fname, sep='\t')
df['sample'] = df.lab_rep.str.rsplit('_', n=1, expand=True)[1]
```

```python
# get tech rep id -- each lab replicate w/ the same sample is a tech rep
temp = df.drop_duplicates()
temp['tech_rep_num'] = temp.sort_values(['sample',
                         'lab_rep'],
                          ascending=[True, True])\
                          .groupby(['sample']) \
                          .cumcount() + 1
temp.sort_values(by=['sample', 'tech_rep_num'])
df = df.merge(temp, how='left', on=['sample', 'lab_rep'])

df['tech_rep'] = df['sample']+'_'+df['tech_rep_num'].astype(str)
```
